```python
import pandas as pd
import numpy as np

#cp - constrictive pericaditis
#trtbps - resting blood pressure
#chol - cholesterol
#fbs - fasting blood sugar
#restecg - resting electrocardiographic
#thalach - maximum heart rate achieved
#exng (exang)- exercise induced angina
#oldpeak - ST depression induced by exercise relative to rest
#slp (slope) - the slope of the peak exercise ST segment
#ca - number of major vessels colored by flourosopy
#thal - thalassemia (body makes abnormal form or inadequate amount of hemoglobin)


#loading csv to read
data = pd.read_csv('heart.csv')

#sort data by age
data = data.sort_values(by = 'age', ascending = False)
print(data)
```

```python
#connection between age and cholesterol

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

plt.title('Connection between age and cholesterol, resting blood pressure and fasting blood sugar')
data.plot(x = 'age', y = ['chol','trtbps','fbs'])
```

```python
#connection between age and cp 
import matplotlib.pyplot as plt

#plt.title('Connection between age and cp')

figura = data.plot(x = 'age', y = ['cp'])
figura.set_title('Connection between age and Constrictive Pericaditis')

```

```python
# Correlation map
f,ax = plt.subplots(figsize=(12,12))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
```

```python
# Line Plot
# color = color, label = label, linewidth = width of line, alpha = opacity
# grid = grid, linestyle = sytle of line

data.restecg.plot(kind = 'line', color = 'g', label = 'Resting Electrocardiographic',linewidth = 1, alpha = 0.5, grid = True, linestyle = ':')
data.trtbps.plot(color = 'r',label = 'Resting Blood Pressure',linewidth = 1, alpha = 0.5, grid = True, linestyle = '-.')
plt.legend(loc='upper right')
```

```python
# Scatter Plot
# x = Age, y = Cholesterol

data.plot(kind = 'scatter',x = 'age', y = 'chol',alpha = 0.5, color = 'red')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.title('Age-Cholesterol Plot')

# x = age , y = restecg

data.plot(kind = 'scatter',x = 'age', y = 'restecg',alpha = 0.5, color = 'blue')
plt.xlabel('Age')
plt.ylabel('restecg')
plt.title('Age-Resting Electrocardiographic Plot')

# x= age, y = trtbps

data.plot(kind = 'scatter',x = 'age', y = 'trtbps',alpha = 0.5, color = 'green')
plt.xlabel('Age')
plt.ylabel('trtbps')
plt.title('Age-Resting Blood Pressure Plot')
```
